# Class to optimize x

```python
class ForX(nn.Module):
    def __init__(self):
        super().__init__()
        self.x = nn.Parameter(torch.Tensor([0.0]))
        self.y = torch.tensor([0.0], requires_grad=True)
    def setY(self, y):
        self.y = torch.tensor(y, requires_grad=True)
    def forward(self):
        f = 2 * torch.square(self.x) + 2 * self.x * self.y + 2 * torch.square(self.y) - 6 * self.x
        return f
```
